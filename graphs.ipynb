# Logistic Regression TF-IDF Results

```python
import matplotlib.pyplot as plt
import pandas as pd

results = pd.read_csv("./results/logistic_regression_tfidf.csv")
iterations = list(range(1, 51))
plt.plot(iterations, results["train accuracy"], label = "train")
plt.plot(iterations, results["test accuracy"], label = "test")

plt.xlabel("Max Iterations")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Max Iterations")
plt.legend()

plt.show()
```

# Compare MLP Models Training

```python
epochs = list(range(1, 11))
mlp_bow_results = pd.read_csv("./results/mlp_bow.csv")
mlp_tfidf_results = pd.read_csv("./results/mlp_tfidf.csv")
mlp_word2vec_results = pd.read_csv("./results/mlp_word2vec.csv")


plt.plot(epochs, mlp_bow_results["loss"], label = "Bag of Words")
plt.plot(epochs, mlp_tfidf_results["loss"], label = "TF-IDF")
plt.plot(epochs, mlp_word2vec_results["loss"], label = "Word2Vec")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Train Loss vs Epochs")
plt.legend()

plt.show()
```

# Compare BERTs
Tokenizer: Max sequence length = 300  
Adam optimizer learning rate = 0.00001  
Batch Size = 32  
Trained for 5 epochs  

BERT:  
Only fine-tuned the classifier

BERT2:  
Froze 6 encoder layers, fine-tune every other layer

```python
epochs = list(range(1, 6))
bert_results = pd.read_csv("./results/bert.csv")
bert2_results = pd.read_csv("./results/bert2.csv")

plt.plot(epochs, bert_results["loss"], label = "BERT")
plt.plot(epochs, bert2_results["loss"], label = "BERT2")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Train Loss vs Epochs")
plt.legend()

plt.show()
```

```python
plt.plot(epochs, bert_results["accuracy"], label = "BERT")
plt.plot(epochs, bert2_results["accuracy"], label = "BERT2")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Train Accuracy vs Epochs")
plt.legend()

plt.show()
```

```python
plt.plot(epochs, bert_results["val_accuracy"], label = "BERT")
plt.plot(epochs, bert2_results["val_accuracy"], label = "BERT2")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Validation Accuracy vs Epochs")
plt.legend()

plt.show()
```
