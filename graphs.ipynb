# Logistic Regression TF-IDF Results

```python
import matplotlib.pyplot as plt
import pandas as pd

results = pd.read_csv("./results/logistic_regression_tfidf.csv")
iterations = list(range(1, 51))
plt.plot(iterations, results["train accuracy"], label = "train")
plt.plot(iterations, results["test accuracy"], label = "test")

plt.xlabel("Max Iterations")
plt.ylabel("Accuracy")
plt.title("TF-IDF Accuracy vs Max Iterations")
plt.legend()

plt.show()
```

# Compare BERTs
Tokenizer: Max sequence length = 300  
Adam optimizer learning rate = 0.00001  
Batch Size = 32  
Trained for 5 epochs  

BERT:  
Only fine-tuned the classifier

BERT2:  
Froze 11 encoder layers, fine-tune every other layer (1 encoder layer)

BERT3:  
Froze 9 encoder layers, fine-tune every other layer (3 encoder layers)

BERT4:  
Froze 6 encoder layers, fine-tune every other layer (6 encoder layers)

BERT5:  
Train entire model (12 encoder layers)

```python
epochs = list(range(1, 6))
bert_results = pd.read_csv("./results/bert.csv")
bert2_results = pd.read_csv("./results/bert2.csv")
bert3_results = pd.read_csv("./results/bert3.csv")
bert4_results = pd.read_csv("./results/bert4.csv")
bert5_results = pd.read_csv("./results/bert5.csv")

plt.plot(epochs, bert_results["loss"], label = "BERT")
plt.plot(epochs, bert2_results["loss"], label = "BERT2")
plt.plot(epochs, bert3_results["loss"], label = "BERT3")
plt.plot(epochs, bert4_results["loss"], label = "BERT4")
plt.plot(epochs, bert5_results["loss"], label = "BERT5")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Train Loss vs Epochs")
plt.legend()

plt.show()
```

```python
plt.plot(epochs, bert_results["accuracy"], label = "BERT")
plt.plot(epochs, bert2_results["accuracy"], label = "BERT2")
plt.plot(epochs, bert3_results["accuracy"], label = "BERT3")
plt.plot(epochs, bert4_results["accuracy"], label = "BERT4")
plt.plot(epochs, bert5_results["accuracy"], label = "BERT5")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Train Accuracy vs Epochs")
plt.legend()

plt.show()
```

```python
plt.plot(epochs, bert_results["val_accuracy"], label = "BERT")
plt.plot(epochs, bert2_results["val_accuracy"], label = "BERT2")
plt.plot(epochs, bert3_results["val_accuracy"], label = "BERT3")
plt.plot(epochs, bert4_results["val_accuracy"], label = "BERT4")
plt.plot(epochs, bert5_results["val_accuracy"], label = "BERT5")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Validation Accuracy vs Epochs")
plt.legend()

plt.show()
```
